```python
import matplotlib.pyplot as plt

uni_mq = excel_data.drop_duplicates(subset = "MQ_KURZNAME")

plt.figure(figsize=(10,8))
plt.scatter(uni_mq["LÄNGE (WGS84)"], uni_mq["BREITE (WGS84)"])
plt.title("MQ Locations in Berlin")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
#when were they set up
```

```python
setup_date = excel_data.groupby("MQ_KURZNAME")["INBETRIEBNAHME"].min()
setup_year = setup_date.dt.year.value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.bar(setup_year.index, setup_year.values)
plt.title("MQ Setup by Year")
plt.xlabel("Year")
plt.ylabel("Number of MQs")
plt.show()
```
